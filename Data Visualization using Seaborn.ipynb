# Data Visualization using Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

Ref- https://seaborn.pydata.org/tutorial.html

```python
import pandas as pd
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].value_counts()
```

```python
sns.scatterplot(x='sepal_length', y="petal_length", data =iris)
```

```python
sns.scatterplot(x='sepal_length', y="petal_length", data =iris, hue="species")
```

```python
sns.scatterplot(x='sepal_length', y="petal_length", data =iris,hue="species", size="species")
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue="species",palette="Set1")
```

```python
sns.lmplot(x='petal_length', y="petal_width", data =iris)
```

```python
sns.lmplot(x='petal_length', y="petal_width", data =iris, hue="species")
```

```python
sns.lmplot(x='petal_length', y="petal_width", data =iris, hue="species" ,markers=['o','*','^'])
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris)
```
